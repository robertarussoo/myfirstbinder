## Esempio 1: una matrice casuale 

````python
import numpy as np
import matplotlib.pyplot as plt

A = np.random.randn(10,15)
print(A[0])
plt.matshow(A)
````

## Esempio 2: altra dimensione e altri colori

````python
import numpy as np
import matplotlib.pyplot as plt

B = np.random.randn(10,40)
print(B[0])
plt.matshow(B,cmap=plt.cm.inferno)
````

## Esempio 3: grafico di una funzione matematica

````python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-2*np.pi,2*np.pi,1000)
Y1 = np.sin(X)
Y2 = np.sin(2*X)
plt.figure(figsize=(20,5))
plt.plot(X,Y1,X,Y2)
````

## Esempio 4: modifichiamo la frequenza

````python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

def graph(n):
    X = np.linspace(-2*np.pi,2*np.pi,1000)
    Y1 = np.sin(X)
    Y2 = np.sin(n*X)
    plt.figure(figsize=(20,5))
    return plt.plot(X,Y1,X,Y2)

interact(graph,n=IntSlider(min=1,max=10,step=1))
````
